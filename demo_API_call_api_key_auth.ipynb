```python
import requests
import base64

from PIL import Image
from io import BytesIO

import uuid
```

```python
invoice_path = r'uctenka02.jpg'

with open(invoice_path, "rb") as fd:
    invoice = fd.read()
    
if invoice_path.split('.')[-1] == 'pdf':
        base64_invoice = base64.encodebytes(invoice)
        base64_invoice_str = base64_invoice.decode('utf-8')
else:
    invoice_imgs = [Image.open(invoice_path)]
    img_byte_arr = BytesIO()
    base64_invoice_str = []
    for img in invoice_imgs:
        img.save(img_byte_arr, format='PNG')
        base64_invoice_str += [base64.encodebytes(img_byte_arr.getvalue()).decode('utf-8') ]
```

```python
my_params = {
    'invoice': base64_invoice_str, 
    'taxDocumentId': str(uuid.uuid4()),
    }

headers = {'x-api-key': 'test-14bb971f-f257-4c6f-8039-4305ebfab2c3'}
```

```python
resp = requests.get('https://nathan-demo.azurewebsites.net/ping', headers=headers)
print(resp)
print(resp.text)
```

```python
resp = requests.post('https://nathan-demo.azurewebsites.net/recognise', json=my_params, headers=headers)
print(resp)
invoice_data = resp.json()
invoice_data
```
